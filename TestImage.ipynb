**Opérations de coût constant :**

- Opérations arithmétiques de base sur des objets codés sur un nombre fini de bits (int ou caractères, pas string ou nd array).
- Opérations de comparaison sur des objets codés sur un nombre fini de bits.
- Transfert de données (lecture et écriture sur un emplacement de mémoire, attention ça marche pour une case d'un nd array, pas lorsqu'on recopie tout ou une partie d'un tableau dans un autre).
- Instructions de contrôle (branchement conditionnel et inconditionnel, appel à une fonction auxiliaire, exponentiation, ...).

Notation $\mathcal{O}$ majore le pire cas, la notation $\Omega$ minore le meuilleur cas.

L'expérience montre que généralement, **le cas moyen est autant proche du pire que du meuilleur cas** , on utilise la notation $\Theta(\alpha_n)$. Ce qui exprime que quelle que soit la configuration d'entrée, le temps d'exécution est grosso-modo proportionnel à $\alpha_n$.

$$ f(n) = \mathcal{O}(\alpha_n) \quad \iff \quad \exists B>0 / \; f(n) \leq B\alpha_n $$

$$ f(n) = \Omega(\alpha_n) \quad \iff \quad \exists A>0 / \; f(n) \geq A \alpha_n $$

$$ f(n) = \Theta(n) \quad \iff \quad \exists A,B / \; B > A > 0, \; A\alpha_n \leq f(n) \leq B \alpha_n  $$

$\mathcal{O}(n \log(n)) : $ coût quasi-linéaire, $\mathcal{O}(\log(n)) : $ coût logarithmique, $\mathcal{O}(n^2) : $ coût quadratique.

**Remarque :** Cf figure, les algorithmes ayant une complexité supérieure à une complexité quadratique sont soit inutilisables, soit utilisables pour de petites voire très petites valeurs de $n$.

## Complexité spatiale :

De la même façon qu’on définit la complexité temporelle d’un algorithme pour évaluer ses performances en
temps de calcul, on peut définir sa complexité spatiale pour évaluer sa consommation en espace mémoire. Le
principe est le même sauf qu’ici on cherche à évaluer l’ordre de grandeur du volume en mémoire utilisé : il
ne s’agit pas d’évaluer précisément combien d’octets sont consommés par un algorithme mais de préciser son
taux de croissance en fonction de la taille n de l’entrée. Cependant, on notera que la complexité spatiale est
bien moins que la complexité temporelle un frein à l’utilisation d’un algorithme : on dispose aujourd’hui le
plus souvent d’une quantité pléthorique de mémoire vive, ce qui rend moins important la détermination de la

```python
arr1 = np.array([[0,1,4], [1,0,0]])
arr2 = np.copy(arr1)
print(id(arr1) == id(arr2))
print(id(arr1[0])==id(arr2[0]))
```

**La fonction `np.copy` crée une copie de premie niveau, pour éviter cela on utilise la fonction `copy.deepcopy()`**

```python
import copy
arr3 = copy.deepcopy(arr1)
print(id(arr1) == id(arr3))
print(id(arr1[0])==id(arr3[0]))
arr3[0][0] = 1
print(arr1[0], arr3[0])
print(id(arr1) == id(arr3))
print(id(arr1[0])==id(arr3[0]))
```
